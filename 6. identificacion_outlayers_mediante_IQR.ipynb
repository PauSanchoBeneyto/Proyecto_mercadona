```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



# Leer el archivo CSV
data = pd.read_csv('supermercado.csv')

# Mostrar las primeras filas del dataset
print(data.head())

# Resumen estadístico
print(data.describe())

# Información del dataset
print(data.info())

# Comprobar valores nulos
print(data.isnull().sum())

# Visualizar la distribución de precios
plt.figure(figsize=(10, 6))
sns.histplot(data['Precios'], bins=30, kde=True)
plt.title('Distribución de Precios')
plt.xlabel('Precios')
plt.ylabel('Frecuencia')
plt.show()

# Visualizar la distribución de precios por unidad
plt.figure(figsize=(10, 6))
sns.histplot(data['Precios_unidad'], bins=30, kde=True)
plt.title('Distribución de Precios por Unidad')
plt.xlabel('Precios por Unidad')
plt.ylabel('Frecuencia')
plt.show()
```

```python
Q1_precios = data['Precios'].quantile(0.25)
Q3_precios = data['Precios'].quantile(0.75)
IQR_precios = Q3_precios - Q1_precios

outliers_precios = data[(data['Precios'] < (Q1_precios - 1.5 * IQR_precios)) |
                        (data['Precios'] > (Q3_precios + 1.5 * IQR_precios))]

print("Outliers en 'Precios':")
print(outliers_precios)

Q1_precios_unidad = data['Precios_unidad'].quantile(0.25)
Q3_precios_unidad = data['Precios_unidad'].quantile(0.75)
IQR_precios_unidad = Q3_precios_unidad - Q1_precios_unidad

outliers_precios_unidad = data[(data['Precios_unidad'] < (Q1_precios_unidad - 1.5 * IQR_precios_unidad)) |
                               (data['Precios_unidad'] > (Q3_precios_unidad + 1.5 * IQR_precios_unidad))]

print("Outliers en 'Precios_unidad':")
print(outliers_precios_unidad)
```

```python
Q1_precios = data['Precios'].quantile(0.25)
Q3_precios = data['Precios'].quantile(0.90)
IQR_precios = Q3_precios - Q1_precios

outliers_precios = data[(data['Precios'] < (Q1_precios - 1.5 * IQR_precios)) |
                        (data['Precios'] > (Q3_precios + 1.5 * IQR_precios))]

print("Outliers en 'Precios':")
print(outliers_precios)

# Calcular el IQR con el percentil 90 para 'Precios_unidad'
Q1_precios_unidad = data['Precios_unidad'].quantile(0.25)
Q3_precios_unidad = data['Precios_unidad'].quantile(0.90)
IQR_precios_unidad = Q3_precios_unidad - Q1_precios_unidad

outliers_precios_unidad = data[(data['Precios_unidad'] < (Q1_precios_unidad - 1.5 * IQR_precios_unidad)) |
                               (data['Precios_unidad'] > (Q3_precios_unidad + 1.5 * IQR_precios_unidad))]

print("Outliers en 'Precios_unidad':")
print(outliers_precios_unidad)
```

```python
# Crear una nueva columna que indique "error" para precios superiores a 600
data['Precio_Estatus'] = data['Precios'].apply(lambda x: 'error' if x > 600 else 'ok')

# Filtrar los registros con "error"
error_data = data[data['Precio_Estatus'] == 'error']

# Guardar los nombres en una lista
error_nombres = error_data['Nombre'].tolist()

# Mostrar los registros con "error"
print(error_data)

# Mostrar la lista de nombres
print(error_nombres)
```
