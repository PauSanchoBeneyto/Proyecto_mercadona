```python
import pandas as pd
```

```python
carrefour = pd.read_csv("Carrefour.csv")
dia = pd.read_csv("Dia.csv")
mercadona = pd.read_csv("Mercadona.csv")
```

```python
carrefour['Supermercado'] = 'Carrefour'
mercadona['Supermercado'] = 'Mercadona'
dia['Supermercado'] = 'Dia'
```

```python
carrefour.head(1)
```

```python
carrefour['Marca']=pd.NA
```

```python
mercadona.head(1)
```

```python
dia.head(1)
```

```python
carrefour_ = carrefour.rename(columns={'nombre': 'Nombre', 'Precios': 'Precios', 'Precios_unit': 'Precios_unidad', 'categorias': 'cat_falsa',
                                       'subcategoria': 'Subcategoria','Categoria_principal': 'Categorias',
                                       'Unidad de referencia': 'Unidad de referencia', 'marcas':'Marca'})

mercadona_ = mercadona.rename(columns={'Nombre': 'Nombre', 'Precio': 'Precios', 'Precio_unidad': 'Precios_unidad',
                                       'subcategoria': 'Categorias','categoria_producto': 'Subcategoria',
                                       'Unidad_medida': 'Unidad de referencia', 'Marca':'Marca'})

dia_ = dia.rename(columns={'Nombre': 'Nombre', 'Precio': 'Precios', 'Precio Unitario': 'Precios_unidad',
                                       'categoria_principal': 'Categorias','Categoria2': 'Subcategoria',
                                       'UM': 'Unidad de referencia', 'Marca':'Marca'})
```

```python
result = carrefour_.merge(mercadona_, on=['Nombre', 'Precios','Precios_unidad','Categorias','Subcategoria','Unidad de referencia','Marca','Supermercado'], how='outer').merge(dia_, on=['Nombre', 'Precios','Precios_unidad','Categorias','Subcategoria','Unidad de referencia','Marca','Supermercado'], how='outer')
```

```python
supermercado = result.drop(['ID_cat', 'ID_subcat','ID_pro','Fecha_actual','Categoria1','cat_falsa'], axis=1)
```
