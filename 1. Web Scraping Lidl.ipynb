#TODAS_LAS_OFERTAS

##Todos los enlaces

```python
url = 'https://www.lidl.es/es/todas-las-ofertas/s2597'
html=req.get(url).text
soup=bs(html, 'html.parser')
```

```python
lista_enlaces=[]
for enlace in soup.find_all('a'):
    link = enlace.get('href')
    lista_enlaces.append(link)

limpieza_anterior = lista_enlaces.index('todas-las-ofertas/s2597')
limpieza_posterior=lista_enlaces.index('/es/newsletter',)

# Eliminamos los elementos hasta el índice encontrado
lista_enlaces = lista_enlaces[limpieza_anterior+2:limpieza_posterior]

# Eliminar duplicados manteniendo el orden
nueva_lista_enlaces = []
enlaces_vistos = set()
for enlace in lista_enlaces:
    if enlace not in enlaces_vistos:
        nueva_lista_enlaces.append(enlace)
        enlaces_vistos.add(enlace)
```

```python
nueva_lista_enlaces
```

##Nombre de las categorias

```python
nombre_categorias=[]
for url in nueva_lista_enlaces:
    parts = url.split('/')
    product_name = parts[-2]  # El nombre del producto está en la penúltima posición
    nombre_categorias.append(product_name)
```

```python
nombre_categorias
```

##Abrir enlace y descargar datos

```python
columns = ['Categoria','Nombre','Precio','Precio_antes']
df_concatenado = pd.DataFrame(columns=columns)
i=0
for enlace in nueva_lista_enlaces:
  chrome_options = Options()
  chrome_options.add_argument('--headless')
  chrome_options.add_argument('--no-sandbox')
  chrome_options.add_argument('--disable-dev-shm-usage')

  ua = UserAgent()
  chrome_options.add_argument(f'user-agent={ua.random}')
  driver = webdriver.Chrome(options=chrome_options)
  driver.get(enlace) # Conexion
  time.sleep(5)
  html=req.get(enlace).text
  soup=bs(html, 'html.parser')

  lista_productos=[]
  lista_precio=[]
  lista_antes=[]
  categoria=[]

  productos = soup.find_all('div', class_="plp-product-grid-box-tile__title")
  precio = soup.find_all('div', class_="plp-product-grid-box-tile__price")

  for producto in productos:
      # Encuentra el elemento <strong> dentro de cada div "plp-product-grid-box-tile__title"
      nombre_producto = producto.find('strong')
      # Accede al texto dentro del elemento <strong> si existe
      if nombre_producto:
          nombre_producto_texto = nombre_producto.text
          lista_productos.append(nombre_producto_texto)
      else:
          lista_productos.append(None)


  for pre in precio:
      # Encuentra el elemento <strong> dentro de cada div "plp-product-grid-box-tile__title"
      precio_producto = pre.find('span')
      # Accede al texto dentro del elemento <strong> si existe
      if precio_producto:
          precio_prod = precio_producto.text
          if '*' in precio_prod:
            # Encuentra la posición del asterisco
            asterisco_pos = precio_prod.index('*')
            # Extrae los tres caracteres previos al asterisco
            precio_producto = precio_prod[asterisco_pos - 5: asterisco_pos]
            lista_precio.append(precio_producto)
          else:
            lista_precio.append(None)

          if 'antes:' in precio_prod:
            # Encuentra la posición del asterisco
            antes_pos = precio_prod.index('antes:')
            # Extrae los tres caracteres previos al asterisco
            precio_antes = precio_prod[antes_pos+15: antes_pos + 19]
            lista_antes.append(precio_antes)
          else:
            lista_antes.append(None)

  longitud = len(lista_precio)
  cat = nombre_categorias[i]
  categoria = [cat] * longitud
  i=i+1

  diccionario={'Categoria': categoria,'Nombre': lista_productos, 'Precio': lista_precio,'Precio_antes': lista_antes}
  df=pd.DataFrame(diccionario)
  df_concatenado = pd.concat([df, df_concatenado], ignore_index=True)
```

```python
fecha_actual = datetime.today().strftime('%Y-%m-%d')  # Formato YYYY-MM-DD

# Agregar una nueva columna con la fecha actual al DataFrame
df_concatenado['Fecha_actual'] = fecha_actual
```

```python
df_aldi=df_concatenado.copy()
```
