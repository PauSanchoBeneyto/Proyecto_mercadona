```python
#Obtener las dos primeras palabras
def dos_primeras_palabras(valor):
    if isinstance(valor, str):
        palabras = valor.split()  # Dividir la cadena en palabras
        return ' '.join(palabras[:2])  # Unir las dos primeras palabras
    return valor


supermercado['marcas'] = supermercado['marcas'].apply(dos_primeras_palabras)

print(supermercado)
```

```python
# Definir la lista de palabras a buscar
palabras_buscadas = ['CARREFOUR', 'NAVIDUL','PROZIS','NESCAFE','BARCELO','BACARDI','HERO','GUMMIES','PROTOS','MAHOU','MARQUES','PARMAREGGIO','GRUYERE','CAMPOMAR','SANDARA','PRINCIPE','TABASCO','WYKE','POZO','DAMM','CHIQUILIN','MARQUES','FONTAMEDA','CHOVI','ACTIVIA','CARRETILLA','PAMPADOUR','CAMPOFRIO','POMPADOUR','CONGUITOS','LACASITOS','SUCHARD','CUCA','CONCA','PRINGLES','BIMBO','SANTIVERI','CIBELES','DULCESOL','VULPI','YOSOY','MILKA','SCHWEPPES','PULEVA','SUNNY','DANONE','OKEY','GALLO','FILIPINOS','LAUKI','RIO','CUIDAT','ROYAL','FUET','REVILLA','SABU','NORDIC']

# Función para modificar las celdas
def buscar_y_modificar(celda, palabras):
    if isinstance(celda, str):  # Verificar si la celda es una cadena
        palabras_encontradas = [palabra for palabra in palabras if palabra in celda]
        if palabras_encontradas:
            return ' '.join(palabras_encontradas)
    return celda

# Aplicar la función a la columna deseada
supermercado['marcas'] = supermercado['marcas'].apply(lambda x: buscar_y_modificar(x, palabras_buscadas))
```
